### Question 3-b.
Encode and decode the message ’my secret message’ (assume ⟨space⟩ $\rightarrow$ ⟨space⟩, . $\rightarrow$ . and , $\rightarrow$ ,).

```python
# message
message = "my secret message"

# create an FSA that accepts the given message
for i in range(len(message)):
    if message[i]==" ":
        print >> compiler, "{0} {1} <space> <space>".format(i,i+1)
    else:
        print >> compiler, "{0} {1} {2} {2}".format(i,i+1,message[i])
print >> compiler, "17"

A = compiler.compile()
A
```

```python
# encodes the message using rot13
encoded_m = fst.compose(A,rot13)

# decodes the encoded message using inverted rot13
decoded_m = fst.compose(encoded_m,rot13.invert())
```

```python
# function that prints the string of the given fst
def print_output(fsa_):
    
    string = fsa_.project(project_output=True).text().split('\n')
    string_ = []
    for i in range(len(string)-2):
        if string[i].split('\t')[2] == '<space>':
            string_.append(" ")
        else:
            string_.append(string[i].split('\t')[2])

    return ''.join(string_)
```

### Question 3-c.
We wish to decipher the message that can be found in the file **4.encoded1.fst** knowing that in order to do so, we must simultaneously allow two transductions, namely *rot13* and *rot16* (a $\rightarrow$ q , b $\rightarrow$ r , ...), so that a can either encode an original `n` or `q`. Build a suitable decoding transducer, apply it to the encoded message and examine the resulting FST (after projecting onto the output symbols). How many states and arcs does it contain? How many states and arcs after removing epsilons, determinizing and minimizing it? How many distinct strings does it represent?

```python
# encoded message
encoded1 = fst.Fst.read("files/4.encoded1.fst")
```

```python
# decoding transducer - simultaneously allow rot13 and rot16
rot1316 = rot13.copy().union(rot16).closure().rmepsilon()
```

```python
# decoded message
decoded1 = fst.compose(encoded1, rot1316).project(project_output=True)
```

```python
# this function counts the number of states and arcs of the given FST
def state_and_arc(fst):
    n_state = 0 ; n_arc = 0
    for state in fst.states():
        n_state += 1
        for arc in fst.arcs(state):
            n_arc += 1
            
    print('no. of states:\t{}'.format(n_state))
    print('no. of arcs:\t{}'.format(n_arc))
```

```python
# encoded message
state_and_arc(encoded1)
```

```python
# decoded message
state_and_arc(decoded1)
```

```python
# decoded message - after epsilon removal, determinization and minimization
decoded1_rdm = fst.determinize(decoded1.rmepsilon()).minimize()
```

```python
state_and_arc(decoded1_rdm)
```

Each non-trivial letters (letters excluding space and punctuations) can be mapped to two different letters. The number of non-trivial letters can be inferred from the increase in the number of arcs in the decoded FST, compared to the encoded FST (649-365=284). Therefore, the decoded FST represents $2^{284}$ distinct strings.

### Question 2-e.
Similarly, a second text belonging to the same novel has been encoded in the file **4.encoded2.fst** by applying the *rot13* cipher and allowing some pairs of consecutive letters to be swapped (excluding ⟨space⟩ or punctuation). Build a suitable decoding transducer2, apply it to the encoded message and compose your answer with the language model. What was the original text?

```python
# this transducer allows swaps between consecutive letters
letters = list("abcdefghijklmnopqrstuvwxyz")

# map the space and punctuation to itself
print >> compiler, "0 53 <space> <space>"
print >> compiler, "0 53 . ."
print >> compiler, "0 53 , ,"

# for each letter,
for i in range(0,26):

    # map the letter to itself (so that the letter can remain in its place)
    print >> compiler, "0 53 {0} {0}".format(letters[i])
    
    # make a new state & map the first letter to an epsilon
    print >> compiler, "0 {0} {1} <eps>".format(i+1, letters[i])
    
    # make another state & map every possible second letter to itself
    for j in range(0,26):
        if j != i:
            print >> compiler, "{0} {1} {2} {2}".format(i+1,i+27,letters[j])

    # map an epsilon back to the first letter / the letter are now swapped
    print >> compiler, "{0} 53 <eps> {1}".format(i+27, letters[i])
            
print >> compiler, "53"

swap_letters = compiler.compile()
```

```python
# mini-version to help visualizing the result
letters = list("abc")
for i in range(0,3):    
    print >> compiler, "0 53 {0} {0}".format(letters[i])
    print >> compiler, "0 {0} {1} <eps>".format(i+1, letters[i])
    for j in range(0,3):
        if j != i:
            print >> compiler, "{0} {1} {2} {2}".format(i+1,i+27,letters[j])
    print >> compiler, "{0} 53 <eps> {1}".format(i+27, letters[i])
print >> compiler, "53"

compiler.compile()
```

```python
# read the given encoded text
encoded2 = fst.Fst.read("files/4.encoded2.fst")
```

```python
# create a transducer that allows rot13 and letter-swap
transducer2 = fst.compose(rot13, swap_letters.arcsort().closure())
```

```python
# decode the text and apply epsilon removal, determinization and minimization
decoded2 = fst.compose(encoded2, transducer2).project(project_output=True)
decoded2_rdm = fst.determinize(decoded2.rmepsilon()).minimize()
```

```python
# composed the decoded FST with lm to find what was the original text
result = fst.compose(decoded2_rdm, lm).project(project_output=True)
result = result.set_output_symbols(ST)
```

#### Step 3: Remove the weight, change the FST into log semiring, and apply the shortest distance operation

The shortest distance operation returns the following

>$$d[q] = \oplus_{p \in P(q,F)} w(p)$$

In case of log semiring, the sum operation is the log sum

>$$d[q] = -\log \left( \sum_{p \in P(q,F)} e^{-w(p)} \right)$$

If the weights are removed (i.e. set to zero), the term $e^{-w(p)}$ would be $1$. Hence, the summation over all available paths would simply yield the number of paths.

>$$d[q] = -\log \left( \sum_{p \in P(q,F)} 1 \right) = -\log (|P(q,F)|)$$

Therefore, the number of distinct paths is given as,

>$$|P(q,F)| = e^{-d[q]}$$

```python
tmp = fst.arcmap(countme_rdm, map_type='rmweight')
print fst.shortestdistance(fst.arcmap(tmp, map_type='to_log'), reverse=True)[0]
```
